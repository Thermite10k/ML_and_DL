```python
def star_converter(row):
    points = row.points
    if(points >= 95 ):
        row.points = 3
    if(points >= 85 & points <95):
        row.points = 2
    if (points < 85):
        row.points = 1
    return row

reviews.apply(star_converter, axis="columns")

```
