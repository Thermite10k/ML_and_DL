```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(iowa_model.predict(X), y)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)

new_model = DecisionTreeRegressor()
new_model.fit(train_X, train_y)

new_predictions = new_model.predict(val_X)
mean_absolute_error(new_predictions, val_y)
```

```python
def get_mea(max_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(random_state = 1, max_leaf_nodes = max_nodes)
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    return(mean_absolute_error(predictions, val_y))
    
```

```python
for max_nodes in [5, 50, 500, 5000]:
    mea = get_mea(max_nodes, train_X, val_X, train_y, val_y)
    print("Max nodes: %d \t\t MEA: %d" %(max_nodes, mea))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state = 1)
forest_model.fit(train_X, train_y)
```

```python
forest_predictions = forest_model.predict(val_X)

forest_mea = mean_absolute_error(forest_predictions, val_y)
forest_mea
```
