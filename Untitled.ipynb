```python
import pandas as pd
```

```python
melbourne_path = "./melb_data.csv"
melbourne_data = pd.read_csv(melbourne_path)
melbourne_data.describe()
```

```python
melbourne_data.columns
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
melbourne_data.describe()
```

```python
y= melbourne_data.Price
melbourne_features = ["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
melbourne_model = DecisionTreeRegressor(random_state=8)
melbourne_model.fit(X, y)
```

```python
print(melbourne_model.predict(X.head()))
```

```python
y
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predicted_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_prices)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
new_model = DecisionTreeRegressor(random_state = 1)
new_model.fit(train_X, train_y)
predictions = new_model.predict(val_X)
new_mea = mean_absolute_error(predictions, val_y)
```

```python
new_mea
```

```python
def get_mea(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(random_state = 1, max_leaf_nodes=max_leaf_nodes)
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    return(mean_absolute_error(predictions, val_y))
```

```python
for max_node in [5, 50, 500, 5000]:
    mea = get_mea(max_node, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t MEA: %d" %(max_node, mea))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
forest_predictions = forest_model.predict(val_X)
forest_mea = mean_absolute_error(val_y, forest_predictions)
forest_mea
```
