```python
import pandas as pd
import numpy as np
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

```

```python
original_data = np.random.exponential(size=1000)

min_ = original_data.min()
max_ = original_data.max()

scaled_data = (original_data- min_) / (max_ - min_)
```

```python
fix, ax = plt.subplots(1, 2, figsize=(12, 3))
sns.histplot(original_data, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Scaled data")

plt.show()
# as you can see, the shape stays the same but the range is now from 0 1 
```

```python
normalized_data = stats.boxcox(original_data)

fix , ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Original Data")
sns.histplot(normalized_data, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Normalized data")

plt.show()
```
